# Plot kernel approximation accuracy

```python
approx_types = ['G', 'Gort', 'ROM', 'QMC', 'GQ', 'B']
kernels = ['Arccos 0', 'Arccos 1', 'RBF']
datasets = ['Powerplant', 'LETTER', 'USPS', 'MNIST', 'CIFAR100', 'LEUKEMIA']
```

```python
with open('paper_figures/data/All_KA', 'rb') as f:
    errs = pkl.load(f)
```

```python
fig = plot_errors(errs, datasets, kernels, approx_types, semilogy=True, acc=False, exact=None)
tp.send_figure(fig, 'Kernel approximation results (semilogy).')
```
